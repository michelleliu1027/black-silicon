```python
## import packages
import json 
import os
import pandas as pd
import re
```

```python
## load data
info = []
for paper in os.listdir('elsevier/'):
    with open('elsevier/' + paper) as f:
        data = json.load(f)
        info.append(data)
```

```python
info_df = pd.DataFrame(info)
info_df['originalText'] = info_df['originalText'].str.lower()
links = [link[1]['@href'] for link in [x['coredata']['link'] for x in info]]
info_df['links'] = links
needed_info_df = info_df[['scopus-eid','originalText']]
```

```python
info_df.iloc[5]['links']
```

```python
# get possible table containing info
pattern = '[ ]table [0-9]{1,2}.{1,1500}[0-9][ ]'
tables = {}
for idx, i in enumerate(needed_info_df.originalText):
    tables[needed_info_df['scopus-eid'][idx]]= re.findall(pattern,i)
```

```python
key_words_located = {}
for key, values in tables.items():
    for value in values:
        unit_contition = 'ma/cm2' in value or 'ma cm−2' in value or 'ma·cm−2' in value
        info_condition = 'v oc' in value or 'voc' in value
        if unit_contition and info_condition:
            try:
                key_words_located[key].append(value)
            except:
                key_words_located[key] = []
                key_words_located[key].append(value)
            
        
```

```python
# -*- coding: utf-8 -*-
# @Time    : 2021/12/1 15:40
# @Author  : WAN Yuwei
# @FileName: elsevier_xml.py
# @Email: yuweiwan2-c@my.cityu.edu.hk
# @Github: https://github.com/yuweiwan
# @Personal Website: https://yuweiwan.github.io/
# !pip3 install httpx
import httpx
import json
import time
import os

def scopus_paper_date(paper_doi, apikey, inskey):
    headers = {
        "X-ELS-APIKey": apikey,
        "X-ELS-Insttoken": inskey,
        "Accept": 'application/xml'
    }
    timeout = httpx.Timeout(10.0, connect=60.0)
    client = httpx.Client(timeout=timeout, headers=headers)
    query = "&view=FULL"
    url = f"https://api.elsevier.com/content/article/doi/" + paper_doi
    r = client.get(url)
    return r


# Get document
my_api_key = "1dbdb72723ccca9a72eed4c9b690e075"
my_ins_key = "87573d4520407e43ab4be3ca6c332d5c"


info_lst = []    
with open('elsevier_info.txt') as f:
    for line in f:
        print(line)
        y = scopus_paper_date(line, my_api_key, my_ins_key).text
        info_lst.append(y)

```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(info_lst[2],'xml')
tables = soup.find_all('ce:table')

rows_thread = []
rows_values = []
with open('out_table.csv', 'w+') as outf:
    for table in tables:
        table_name = table.find_all('ce:label')
        ouf.write(table_name+'\n')
        thread_name = table.find_all('thead')
        for i in thread_name:
            rows_thread.append(i.text.strip())
            display(i.text.strip())
            display(i.text.strip().replace('\n',','))
    #         display(pd.DataFrame(i.text.strip()))

    #     print(len(thread_name))
    #     for i in thread_name:
    #         print(i.text)
        tbody_name = table.find_all('tbody')
        for i in tbody_name:
            display(i.text.strip().strip().replace('\n',',').replace(',,,','; ').split('; '))
            rows_values.append(i.text.strip().replace('\n',',').replace(',,,','; ').split('; '))
    #     for i in tbody_name:
    #         print(i.text)
```
