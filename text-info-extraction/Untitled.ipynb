```python
# -*- coding: utf-8 -*-
# @Time    : 2021/12/1 15:40
# @Author  : WAN Yuwei
# @FileName: elsevier_xml.py
# @Email: yuweiwan2-c@my.cityu.edu.hk
# @Github: https://github.com/yuweiwan
# @Personal Website: https://yuweiwan.github.io/
# !pip3 install httpx
import httpx
import json
import time
import os

def scopus_paper_date(paper_doi, apikey, inskey):
    headers = {
        "X-ELS-APIKey": apikey,
        "X-ELS-Insttoken": inskey,
        "Accept": 'application/xml'
    }
    timeout = httpx.Timeout(10.0, connect=60.0)
    client = httpx.Client(timeout=timeout, headers=headers)
    query = "&view=FULL"
    url = f"https://api.elsevier.com/content/article/doi/" + paper_doi
    r = client.get(url)
    return r


# Get document
my_api_key = "1dbdb72723ccca9a72eed4c9b690e075"
my_ins_key = "87573d4520407e43ab4be3ca6c332d5c"


info_lst = []
ids = []
with open('elsevier_info.txt') as f:
    for line in f:
        print(line)
        y = scopus_paper_date(line, my_api_key, my_ins_key).text
        info_lst.append(y)
        ids.append(line)

```

```python
from bs4 import BeautifulSoup
not_working_tables = []
not_working_dfs = []
with open('out_table.csv', 'w+') as outf:
    for paperid in range(len(ids)):
        try:
            outf.write('*'*4 + ids[paperid].strip() + '*'*4 +'\n')
            soup = BeautifulSoup(info_lst[paperid],'xml')
            tables = soup.find_all('ce:table')

            table_names = []
            rows_thread = []
            rows_values = []


            for table in tables:

                table_name = table.find_all('ce:label')
                for i in table_name:
                    table_names.append(i.text)


                thread_name = table.find_all('thead')
                for i in thread_name:
                    rows_thread.append(i.text.strip().replace('\n',','))


                tbody_name = table.find_all('tbody')
                for i in tbody_name:
                    rows_values.append(i.text.strip().replace('\n',',').replace(',,,','; ').split('; '))

            for idx in range(len(table_names)):
#                 print(table_names[idx])
#                 print(rows_thread[idx])
                outf.write(table_names[idx]+'\n')
                outf.write(rows_thread[idx]+'\n')

                for subidx in range(len(rows_values[idx])):
#                     print(rows_values[idx][subidx])
                    outf.write(rows_values[idx][subidx]+'\n')
            outf.write('\n')
        except:
                not_working_tables.append(ids[paperid])
```
