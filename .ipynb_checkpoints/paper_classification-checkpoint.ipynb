# Data Analysis of doi List 
## Preprocess

```python
# every is a list of every article info (by dividing content with '\n')
every = []
# for i in range(35):
#     if i == 0:
filename = 'savedrecs.txt'
#     else:
#         filename = 'savedrecs' + ' (' + str(i) + ').txt'
with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    last_end = 0
    for idx, line in enumerate(lines):
        if line == '\n':
            every.append(lines[last_end:idx])
            last_end = idx+1

pu_doi = {}
doi_dic = {}
for e in every:
    doi = []
    pu = []
    ab = []
    ji = []
    py = []
    dt = []
    sc = []
    tc = []
    wc = []
    i_d = []
    ti = []
    de = []
    cur = []
    for line in e:
        sp = line.split(' ', 1)
        if sp[0] not in ['DI','PU','AB','JI','PY','DT','SC','TC','WC','ID','TI','DE','']:
            cur = []
            continue
        if sp[0] == 'DI':
            doi.append(sp[1][:-1])
            cur = doi
        elif sp[0] == 'PU':
            pu.append(sp[1][:-1])
            cur = pu
        elif sp[0] == 'AB':
            ab.append(sp[1][:-1])
            cur = ab
        elif sp[0] == 'JI':
            ji.append(sp[1][:-1])
            cur = ji
        elif sp[0] == 'PY':
            py.append(sp[1][:-1])
            cur = py
        elif sp[0] == 'DT':
            dt.append(sp[1][:-1])
            cur = dt
        elif sp[0] == 'SC':
            sc.append(sp[1][:-1])
            cur = sc
        elif sp[0] == 'TC':
            tc.append(sp[1][:-1])
            cur = tc
        elif sp[0] == 'WC':
            wc.append(sp[1][:-1])
            cur = wc
        elif sp[0] == 'ID':
            i_d.append(sp[1][:-1])
            cur = i_d
        elif sp[0] == 'TI':
            ti.append(sp[1][:-1])
            cur = ti
        elif sp[0] == 'DE':
            de.append(sp[1][:-1])
            cur = de
        elif sp[0] == '':
#             print(line)
            cur.append(sp[1][:-1])
        else:
            pass
    ti = [' '.join(filter(None, ' '.join(ti).split(' ')))]
    
    # pu
    if len(doi) == 1 and len(pu) == 1:
        if pu[0] not in pu_doi.keys():
            pu_doi[pu[0]] = [doi[0]]
        else:
            pu_doi[pu[0]].append(doi[0])
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['publisher'] = pu[0]
        else:
            doi_dic[doi[0]]['publisher'] = pu[0]
    # ab
    if len(doi)== 1 and len(ab) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['abstract'] = ab[0].lower()
        else:
            doi_dic[doi[0]]['abstract'] = ab[0].lower()
    # ji
    if len(doi)== 1 and len(ji) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['journal'] = ji[0]
        else:
            doi_dic[doi[0]]['journal'] = ji[0]
    # py
    if len(doi)== 1 and len(py) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['year'] = py[0]
        else:
            doi_dic[doi[0]]['year'] = py[0]
    # dt
    if len(doi)== 1 and len(dt) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['type'] = dt[0]
        else:
            doi_dic[doi[0]]['type'] = dt[0]
    # sc
    if len(doi)== 1 and len(sc) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['category'] = sc[0]
        else:
            doi_dic[doi[0]]['category'] = sc[0]
    # tc
    if len(doi)== 1 and len(tc) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['citation'] = tc[0]
        else:
            doi_dic[doi[0]]['citation'] = tc[0]
    # wc
    if len(doi)== 1 and len(wc) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['wc'] = wc[0]
        else:
            doi_dic[doi[0]]['wc'] = wc[0]
    # id
    if len(doi)== 1 and len(i_d) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['i_d'] = i_d[0]
        else:
            doi_dic[doi[0]]['i_d'] = i_d[0]
    # ti
    if len(doi)== 1 and len(ti) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['title']= ti[0].lower()
        else:
            doi_dic[doi[0]]['title'] = ti[0].lower()
    # de
    if len(doi)== 1 and len(de) == 1:
        if doi[0] not in doi_dic.keys():
            doi_dic[doi[0]] = {}
            doi_dic[doi[0]]['keywords'] = de[0]
        else:
            doi_dic[doi[0]]['keywords'] = de[0]
```

```python
def f(x):
    return {key: value for key, value in x.items() if key in ['title','abstract','type']}
info_dic = dict(map(lambda x: (x[0], f(x[1])), doi_dic.items()))
info_dic
```

## Recall as evaluation metrics:
#### Only title

```python
from sklearn.metrics import recall_score
oks = {key:value for key, value in info_dic.items() if ('solar cell' in value['title'] and 'black' in value['title'] and 'silicon' in value['title']) or \
       ('solar cell' in value['title'] and 'nano' in value['title'])}
ok_title_pred = [value['title'] for key, value in oks.items()]

bi_pred_title = [1 if i in ok_title_pred else 0 for i in ok_titles_given]
bi_given = [1] * len(ok_titles_given)
recall_score(bi_given, bi_pred_title)

```

#### Only abstract

```python
def only_abstract(ipt):
    try:
        return ('%' in ipt['abstract'])
    except:
        pass
```

```python
oks_abstract = {key:value for key, value in info_dic.items() if only_abstract(value)}
ok_abstract_pred = [value['title'] for key, value in oks_abstract.items()]
bi_pred_abstract = [1 if i in ok_abstract_pred else 0 for i in ok_titles_given]
recall_score(bi_given, bi_pred_abstract)
```
